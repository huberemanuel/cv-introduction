# Variáveis e seus tipos

Em Python é possível declarar variáveis de diversos tipos, inclusive mudar o tipo de uma variável em execução, pois Python é uma linguagem dinamiacmente tipada.


````
a = "Minha primeira variavel"
print(a)
````

````
a = 10
print(a)
````

Números reais (ponto flutuante) são separados por `.`  e não por `,`

````
a = 10.25
print(a)
````

````
a = 10,25
print(a)
````

````
a, b = 10,25
print(a)
print(b)
````

# Operações condicionais

Os operadores `if`, `elif` e `else` permitem a execução de operações condicionais em código Python

````
# Verificar se o usuário é maior de idade
idade = int(input("Digite a sua idade: "))
if idade >= 18:
    print("Usuário é maior de idade")
else:
    print("Usuário é menor de idade")
````

````
# Múltiplas condições
idade = int(input("Digite a sua idade: "))
if idade >= 60:
    print("Usuário é idoso")
elif idade >= 18:
    print("Usuário é maior de idade")
else:
    print("Usuário é menor de idade")
````

# Funções

Funções permitem o reuso de um trecho de código específico, deixando o código mais conciso.

````
def calc_imc(height, weight):
    return weight / (height * height)

imc_score = calc_imc(1.91, 90)
print(imc_score)
````

````
def calc_imc_level(imc_score):
    if imc_score >= 40:
        return "Obesidade grave"
    elif imc_score >= 30:
        return "Obesidade"
    elif imc_score >= 25:
        return "Sobrepreso"
    elif imc_score >= 18.5:
        return "Normal"
    else: 
        return "Magreza"

print(calc_imc_level(imc_score))
````

# Loops

Loops são estruturas de repetição que executam o trecho de código até uma condição de parada ter sido alcançada

````
for i in range(10):
    print(i)
````

````
for i in range(10, 20):
    print(i)
````

````
# Iterando com passo 2
for i in range(0, 10, 2):
    print(i)
````

# Listas, Dicionários, Tuplas e Classes

Além dos tipos básicos de variáveis, Python também suporta estruturas mais complexas que facilitam as operações mais elaboradas.

## Listas

````
scores = [0, 2, 4, 10]
for i in range(len(scores)):
    print("Nota: ", scores[i])
````

````
for score in scores:
    print("Nota: ", score)
````

````
mean = 0
for score in scores:
    mean += score
mean /= len(scores)
print(mean)
````

````
total_scores = sum(scores)
print(total_scores)
````

````
mean = sum(scores) / len(scores)
print(mean)
````

````
doubled_scores = [2 * x for x in scores]
print(doubled_scores)
````

````
doubled_scores * 2
````

## Dicionários

Dicionários são estruturas de dados baseadas em chave e valor, permitindo um acesso eficiente.

````
students = {}
````

````
students_scores = {
    "Aluno 1": 10.0,
    "Aluno 2": 3.0,
    "Aluno 3": 5.0,
    "Aluno 4": 8.0,
}

print(students_scores["Aluno 1"])
````

Os acessos são feitos pela chave (primeiro elemento), diferente de listas onde o acesso é feito pelo índice de cada elemento.

````
for key, value in students_scores.items():
    print(key, value)
````

Os valores de dicionários podem ser outros dicionários

````
students = {
    "Student 1": {
        "id": 1,
        "age": 10
    },
    "Student 2": {
        "id": 2,
        "age": 15
    },
    "Student 3": {
        "id": 3,
        "age": 25
    }
}

for key, value in students.items():
    print("Student name: ", key)
    print("Student id: ", value["id"])
    print("Student age: ", value["age"])
````

## Tuplas

Uma tupla é similar a uma lista, porém os seus valores são imutáveis

````
ages = (10, 1, 2, 3, 4) # Definidas utilizando parênteses
print(ages)
````

````
for age in ages:
    print(age)
````

````
ages[0] = 1
````

## Classes

As classes permitem criar objetos com propridades, que facilitam o uso de estruturas mais complexas

````
class Student():
    name = "Aluno 1"
    scores = [10, 2, 9, 7]

student1 = Student()
print(student1.name)
print(student1.scores)
````

Além de propriedades, classes também podem possuir métodos:

````
class Student():
    name = "Aluno 1"
    scores = [10, 2, 9, 7]

    def mean(self):
        return sum(self.scores) / len(self.scores)

student1 = Student()
print(student1.mean())
````

````
# Escrever em arquivo

with open("input.txt", "w") as input_file:
    input_file.write("Meu primeiro arquivo de texto")
````
