# O que são Tensores? 

Tensor é uma generalização de matrizes e vetores, onde o que os diferencia é o seu **rank**, ou número de dimensões.

Tensor de rank 1: escalar (Exemplo: 0)

Tensor de rank 2: vetor (Exemplo: [1, 2, 3, 4])

Tensor de rank 3: matriz (Exemplo: [[1, 2], [2, 3])

O Tensor é o bloco básico do Tensorflow, todas as operações realizadas com a biblioteca utilizam Tensores.

Verificar se o tensorflow está executando as operações de forma dinâmica (Eager Execution)

```
tf.executing_eagerly() # Forma padrão de execução
```

# Operações matemáticas


```
x = tf.constant([1, 2, 3])
y = tf.constant([9, 8, 7])
```

Algumas operações possuem suporte a tensores com dimensões diferentes (broadcasting)

```
a = tf.constant([2, 3])
b = tf.constant(6)
a + b
```

```
a = tf.constant([2, 3])
b = tf.constant([6, 3, 9])
a + b
```

Subtração de tensores

```
tf.subtract(x, y), x - y
```

Divisão de tensores

# tf.constant vs tf.Variable

Tensores criados a partir da função `tf.constant` não podem ser modificados, pois são constantes. Para criar tensores que possam ser modificados, utilize a `tf.Variable`

```
v = tf.Variable([3, 2, 1])
v[1].assign(7)
v
```

```
tf.Variable([False, True, False])
```

# Compatibilidade com Numpy

Tensorflow 2 já possui compatibilidade com numpy, isso signifca que o usuário pode criar tensores a partir de objetos `numpy.array`

```
import numpy as np

tf.constant(np.zeros((2, 2)), dtype=tf.float32)
```

# Exercícios

Crie um Tensor unidimensional com shape (20) com todos os valores iguais a `False`


```
v_ori = tf.Variable([False] * 20)
v_ori
```

Utilizando o vetor anterior, altere o valor para `True` nas posições pares.

Exemplo: 

v_ori = [False, False, False, False]

v_new = [True, False, True, False]

```
for i in range(v_ori.shape[0]):
    if i % 2 == 0:
        v_ori[i].assign(True)
v_ori
```

Solução mais elegante, ainda não suportada pelo TF 2

```
import numpy as np
v = np.array([False] * 20)
v[::2] = True
v
```

Crie um Tensor bidimensional com shape (10, 10) com todos os valores zerados, em seguida preencha a subregião indexada por [2:7,2:4] com o valor 1.



```
my_tensor = tf.Variable(tf.zeros((10, 10)))
my_tensor
```

```
for i in range(2, 8):
    for j in range(2, 5):
        my_tensor[i, j].assign(1)
my_tensor
```

Solução elegante não suportada pelo Tensorflow

```
import numpy as np
v = np.zeros((10, 10))
v[2:8,2:5] = 1
v
```
